```python
%matplotlib inline
from math import isnan as nan
import matplotlib.pyplot as plt
import pandas
import numpy as np
import seaborn as sns
```

```python
df = pandas.read_csv('data.csv')
left_x = df['LGazePointOnDisplayX']
left_y = df['LGazePointOnDisplayY']
right_x = df['RGazePointOnDisplayX']
right_y = df['RGazePointOnDisplayY']
left = []
right = []

#Eliminate NaN Values
for lx, ly, rx, ry in zip(left_x, left_y, right_x, right_y):
	if not (nan(lx) or nan(ly) or nan(rx) or nan(ry)):
		left.append((lx, ly))
		right.append((rx, ry))
        
#isolate variables by left x, left y, right x, and right y
coors = zip(left, right)
lx, ly = zip(*left)
rx, ry = zip(*right)
```

```python
# Journey plotting - left eye
left_plot = plt.plot(lx, ly, label='Left Eye Position')
```

```python
#Journey plotting - right eye
right_plot = plt.plot(rx, ry, label='Right Eye Position')
```

```python
#Overlay plotting - Blue = Left Eye, Orange = Right Eye
left_plot = plt.plot(lx, ly, label='Left Eye Position', alpha=.7)
right_plot = plt.plot(rx, ry, label='Right Eye Position', alpha=.5)
```

```python
#Overlay, but scattered
left_plot = plt.scatter(lx, ly, label='Left Eye Position', alpha=.7)
right_plot = plt.scatter(rx, ry, label='Right Eye Position', alpha=.5)
```
